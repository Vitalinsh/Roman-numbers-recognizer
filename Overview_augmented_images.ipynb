```python
import os
import math

import matplotlib.pyplot as plt

from utils import load_data_to_mem, augmentation
```

```python
main_dir = "."
data_dir = os.path.join(main_dir, "data")
train_dir = os.path.join(data_dir, "train")

classes = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII']
```

```python
X, y = load_data_to_mem(train_dir, classes)
print("Before augmantation len(X) =", len(X), ", len(y) =", len(y))
X, y = augmentation(X, y)
print("After augmantation len(X) =", len(X), ", len(y) =", len(y))
```

```python
def show_images(images, labels):
    nb_cols = 3
    nb_rows = math.ceil(len(images) / nb_cols)

    fig, axs = plt.subplots(nb_rows, nb_cols, figsize=(10, 10))

    n = 0
    for i in range(0, nb_rows):
        for j in range(0, nb_cols):
            if n < len(images):
                axs[i, j].xaxis.set_ticklabels([])
                axs[i, j].yaxis.set_ticklabels([])
                axs[i, j].set_title(labels[n], fontsize=15)
                axs[i, j].imshow(images[n], cmap="gray")
                n += 1
                
```

```python
imgs_to_show = [X[1000], X[2000], X[3000],
                X[4000], X[5000], X[6000],
                X[7000], X[8000], X[9000]]
labels_to_show = [y[1000], y[2000], y[3000],
                  y[4000], y[5000], y[6000],
                  y[7000], y[8000], y[9000]]

show_images(imgs_to_show, labels_to_show)
```
