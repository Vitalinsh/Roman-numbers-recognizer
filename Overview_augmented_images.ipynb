```python
import os

from utils import load_data_to_mem, augmentation, show_images
```

```python
main_dir = "."
data_dir = os.path.join(main_dir, "data")
train_dir = os.path.join(data_dir, "train")

classes = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII']
```

```python
X, y = load_data_to_mem(train_dir, classes, img_height=64, img_width=64)
print("Before augmantation len(X) =", len(X), ", len(y) =", len(y))
X, y = augmentation(X, y, n_transform=5)
print("After augmantation len(X) =", len(X), ", len(y) =", len(y))
```

```python
idx = [200, 349, 555, 650, 800, 929]

n = 5
for i in range(5):
    for j in range(6):
        idx.append(idx[n - 5] + 940)
        n += 1
imgs_to_show = [X[i] for i in idx]
labels_to_show = [y[i] for i in idx]

show_images(imgs_to_show, labels_to_show)
```
